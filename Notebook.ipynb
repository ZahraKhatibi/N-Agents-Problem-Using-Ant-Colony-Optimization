```python
n = 100
best_cost = 350
evaporation_rate = 0.1
alfa = 1
beta = 7
ant_number = 5
best_ant_cost = 10e17
best_ant_path = [-1 for i in range(n)]
```

```python
f = open("job2.assign")
m = f.read().split("\n")
data = [[0 for _ in range(n)] for _ in range(n)]
for i in range(0,n):
    data[i] = m[i].split()
```

```python
pheromone = [[1 for i in range(n)] for j in range(n)]
```

```python
class ANT:
    def __init__(self):        
        self.path = []
        self.cost = 0
        self.tasks = pd.Series([True]*n)
```

```python
def probability(row,domain):
    prob = [0 for i in range(0,len(domain))]
    suum = 0
    for i in range(0,len(domain)):
        prob[i] += (pheromone[row][domain[i]]**alfa) * ((1/int(data[row][domain[i]]))**beta)
        suum += (pheromone[row][domain[i]]**alfa) * ((1/int(data[row][domain[i]]))**beta)
    for i in range(0,len(domain)):
            prob[i] /= suum
    return prob
```

```python
import numpy as np
import pandas as pd
def creatPath(Ant):
    size = len(Ant.path)
    while(len(Ant.path)!= n):
        x = np.random.choice(Ant.tasks[Ant.tasks==True].index , p = probability(size, Ant.tasks[Ant.tasks==True].index))
        Ant.path.append(x)
        Ant.cost += int(data[size][x])
        Ant.tasks[x] = False
        size+=1
    global best_ant_cost
    global best_ant_path
    if(Ant.cost<best_ant_cost):
        best_ant_path = Ant.path
        best_ant_cost = Ant.cost
    return Ant
```

```python
def pheromone_maker(Ant):
    for i in range(0,n):
        pheromone[i][Ant.path[i]] += (best_cost / int(data[i][Ant.path[i]]))/100
```

```python
def evaporation():
    for i in range(0,n):
        for j in range(0,n):
            pheromone[i][j] *= (1-evaporation_rate)
            pheromone[i][j]  = max(pheromone[i][j],1)
```

```python
while(best_ant_cost> best_cost-2):
    for i in range(ant_number):
        ant = ANT()
        creatPath(ant)
        pheromone_maker(ant)
        evaporation()
    print("COST: ", best_ant_cost)
    print("PATH: " , best_ant_path)
    print("------------")
    
```
